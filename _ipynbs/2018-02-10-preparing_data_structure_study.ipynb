# 자료구조를 배우기 위한 준비

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1-5 섬 관광 프로그램

class Node:
    def __init__(self, name, left=None, right=None):  # 섬 생성자
        self.name = name
        self.left = left
        self.right = right
        
def map():  # 지도 만들기
    n1 = Node('H')
    n2 = Node('F')
    n3 = Node('S')
    n4 = Node('U')
    n5 = Node('E')
    n6 = Node('Z')
    n7 = Node('K')
    n8 = Node('N')
    n9 = Node('A')
    n10 = Node('Y')
    n11 = Node('T')
    n1.left = n2
    n1.right = n3
    n2.left = n4
    n2.right = n5
    n3.left = n6
    n3.right = n7
    n4.left = n8
    n5.left = n9
    n7.right = n10
    n9.right = n11
    return n1  # 시작 섬 리턴

def A_course(n):
    if n != None:
        print(n.name, '->', end=' ')
        A_course(n.left)
        A_course(n.right)
        
def B_course(n):
    if n != None:
        B_course(n.left)
        print(n.name, '->', end=' ')
        B_course(n.right)
        
def C_course(n):
    if n != None:
        C_course(n.left)
        C_course(n.right)
        print(n.name, '->', end=' ')
        
start = map()
print('A코스:\t', end='')
A_course(start)
print('\nB코스:\t', end='')
B_course(start)
print('\nC코스:\t', end='')
C_course(start)
```

### 1.22 다음 함수에 대해 abc(78)를 호출한 결과는?

```python
def abc(n):
    r = n % 2
    print('*', end=' ')
    if n >= 2:
        abc(n//2)
    print('%d' % r, end=' ')
    return

abc(78)
```

### 1.23 다음 함수에 대해 test('110100111', 4)가 리턴하는 값은?

```python
def test(s, last):
    if last < 0:
        return 0
    if s[-1] == '0':
        return 2 * test(s, last-1)
    return 1 + 2 * test(s, last-1)

test('110100111', 4)
```
